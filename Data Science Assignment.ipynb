```python
attack_mean = data["Attack"].mean()

def set_attack(val):
    if val > attack_mean:
        return "Attack High"
    elif val == attack_mean:
        return "Attack Neutral"
    else:
        return "Attack Low"
```

```python
data["Attack_high_low"] = data["Attack"].apply(set_attack)
data
```

```python
 hp_mean = data["HP"].mean()
    
def set_hp(val):
    if val > hp_mean:
        return "hp High"
    elif val == hp_mean:
        return "hp neutal"
    else:
        return " hp low"
```

```python
data["HP_high_low"] = data["HP"].apply(set_hp)
data
```

```python
data["HP"].mean()
```

```python
speed_mean = data["Speed"].mean()
    
def set_speed(val):
    if val > speed_mean:
        return "speed High"
    elif val == speed_mean:
        return "speed neutal"
    else:
        return " speed low"
```

```python
data["Speed_high_low"] = data["Speed"].apply(set_speed)
data
```
